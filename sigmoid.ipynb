```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x_vals = np.linspace(-10, 10, 200)
y_vals = sigmoid(x_vals)

points = np.array([-7, -3, 3, 6])  # x locations
labels = np.array([0, 0, 1, 1])    # y = class labels (0 or 1)

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, ax in enumerate(axes):
    ax.plot(x_vals, y_vals, label="Sigmoid", color="blue")
    
    # Mark all points
    ax.scatter(points, labels, c=labels, cmap="bwr", s=100, edgecolors="k")
    
    # Highlight one point per subplot for clarity
    ax.scatter(points[i], labels[i], c="gold", s=200, marker="*", label="Focus Point")
    
    ax.axhline(0.5, color="gray", linestyle="--", linewidth=1)
    ax.set_title(f"Sigmoid with point x={points[i]}, class={labels[i]}")
    ax.set_xlabel("x")
    ax.set_ylabel("Predicted probability")
    ax.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
