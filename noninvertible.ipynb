```python
import numpy as np
import plotly.graph_objects as go

# Sample data
X = np.array([[1,2],[2,4],[3,6],[4,8]])
y = np.array([5,9,13,17])

# Add intercept column
X_with_intercept = np.hstack([np.ones((X.shape[0],1)), X])

# Ridge regression coefficients
lambda_ridge = 0.1
I = np.eye(X_with_intercept.shape[1])
beta_ridge = np.linalg.inv(X_with_intercept.T @ X_with_intercept + lambda_ridge * I) @ X_with_intercept.T @ y

print("Ridge coefficients:", beta_ridge)

# Create a grid for plotting the regression plane
x1_range = np.linspace(0,5,20)
x2_range = np.linspace(0,10,20)   # broader than 2*x1_range so it's general
x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)

# Predicted values on the grid
y_mesh = beta_ridge[0] + beta_ridge[1]*x1_mesh + beta_ridge[2]*x2_mesh

# Create 3D plot
fig = go.Figure()

# Scatter plot of original data
fig.add_trace(go.Scatter3d(
    x=X[:,0], y=X[:,1], z=y,
    mode='markers',
    marker=dict(size=6, color='red'),
    name='Data points'
))

# Regression plane
fig.add_trace(go.Surface(
    x=x1_mesh, y=x2_mesh, z=y_mesh,
    colorscale='Viridis',
    opacity=0.6,
    name='Ridge plane'
))

# Labels
fig.update_layout(
    scene=dict(
        xaxis_title='X1',
        yaxis_title='X2',
        zaxis_title='y'
    ),
    title="Ridge Regression Plane in 3D"
)

fig.show()
```
